# 2. Customization
## (1) Labels

```python
# Basic scatter plot, log scale
plt.scatter(train.GrLivArea ,train.SalePrice)
plt.xscale('log') 

# Strings
xlab = 'train.GrLivArea'
ylab = 'SalePrice'
title = 'log scale Relation'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)


# Add title
plt.title(title)
```

```python
# Basic scatter plot, log scale
plt.scatter(train.GrLivArea ,train.SalePrice)

# Previous customizations
plt.xscale('log') 
xlab = 'train.GrLivArea'
ylab = 'SalePrice'
title = 'log scale Relation'

# Add axis labels

tick_val = [100,1000,10000]
tick_lab = ['0.1k','1k','10k']

plt.xticks(tick_val,tick_lab)

plt.show()
```

# (3) Sizes


```python
# Store pop as a numpy array: np_pop


# Double np_pop
pop = np.roots(train.GrLivArea)
np_pop = np.array(pop)
np_pop = np_pop * 2


# Update: set s argument to np_pop
#pop = np.roots(train.GrLivArea)
plt.scatter(train.GrLivArea ,train.SalePrice, s=np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('train.GrLivArea')
plt.ylabel('SalePrice')
plt.title('log scale Relation')
plt.xticks([100,1000,10000], ['0.1k','1k','10k'])

# Display the plot
plt.show()
```
